## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
resp = r.text

resp = json.loads(resp)
resp[0].keys()
```

```python
resp
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for x in resp:
    print(x['title']+' / '+ x['createdAt']+' / '+str(x['commentCount'])+' / '+str(x['likeCount']))
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
# 熱門文章
comment = [x['commentCount'] for x in resp]
like = [x['likeCount']for x in resp]
print('平均留言人數：%.2f' %(sum(comment)/len(comment)))
print('平均按讚人數：%.2f' %(sum(like)/len(like)))
```

# 非熱門

```python
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
resp = r.text

resp = json.loads(resp)
comment = [x['commentCount'] for x in resp]
like = [x['likeCount']for x in resp]
print('平均留言人數：%.2f' %(sum(comment)/len(comment)))
print('平均按讚人數：%.2f' %(sum(like)/len(like)))
```
