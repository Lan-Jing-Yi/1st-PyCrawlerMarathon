```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
resp = r.text

resp = json.loads(resp)
resp[0].keys()
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
for x in resp:
    print(x['title']+' / '+ x['createdAt']+' / '+str(x['commentCount'])+' / '+str(x['likeCount']))
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
# ç†±é–€æ–‡ç« 
comment = [x['commentCount'] for x in resp]
like = [x['likeCount']for x in resp]
print('å¹³å‡ç•™è¨€äººæ•¸ï¼š%.2f' %(sum(comment)/len(comment)))
print('å¹³å‡æŒ‰è®šäººæ•¸ï¼š%.2f' %(sum(like)/len(like)))
```
