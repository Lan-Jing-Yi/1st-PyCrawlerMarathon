## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook

browser = webdriver.Chrome(executable_path='/Users/lanjingyi/Desktop/chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

import time
for i in tqdm_notebook(range(50)):
    time.sleep(5)
    browser.execute_script("window.scrollTo(0, 20000);")
    
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)

```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import datetime
for d in soup.find(class_="part_list_2").find_all('h3'):
    times = datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M')
    if times > datetime(2020,2,13,20,30):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import datetime
for d in soup.find(class_="part_list_2").find_all('h3'):
    times = datetime.strptime(d.find(class_="date").text,'%Y/%m/%d %H:%M')
    if  datetime(2020,2,13,17,0) > times > datetime(2020,2,13,15,0):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
