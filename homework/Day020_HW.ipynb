```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python

d = dict()
for ele in table.find_all('tr'):
    if len(ele.find_all('td')) == 5:
        air_kind = ele.find_all('td')[0].string
        d[air_kind] = dict()
        d[air_kind][ele.find_all('td')[2].string] = ele.find_all('td')[2].string

        print(ele.find_all('td')[0].string)
        print(ele.find_all('td')[2].string)
        print(ele.find_all('td')[3].string)
        print('---------------------')
    elif len(ele.find_all('td')) == 3 :
        d[air_kind][ele.find_all('td')[0].string] = ele.find_all('td')[1].string
        print(ele.find_all('td')[0].string)
        print(ele.find_all('td')[1].string)
        print('---------------------')

```
