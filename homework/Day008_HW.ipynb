### 1. Dcard 網址： https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

r = requests.get(url)

r.encoding = 'utf-8'
print(r.text[0:3000])
```

```python
print('Request 取回之後該怎麼取出資料，資料型態是什麼？ =>')

'''
string 格式
'''
```

```python
print('為什麼要使用 BeatifulSoup 處理？處理後的型態是什麼？ => ')

'''
用BeatifulSoup 可以找到我們需要的標籤內容
'''

content = BeautifulSoup(r.text, "lxml")
content
```

### 2. 知乎： https://www.zhihu.com/explore

```python
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'

print(r.text[0:600])
```

### 3. 觀察一下知乎回來的資料好像有點怪怪的，該怎麼解決？

```python
import requests
url = 'https://www.zhihu.com/explore'

'''
用header 做偽裝
'''
headers = {'Referer':'https://accounts.pixiv.net/loginlang=zh&source=pc&view_type=page&ref=wwwtop_accounts_index',#如某些網站（如p站）要檢查referer，就給他加上
'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.146 Safari/537.36'#每個爬蟲必備的偽裝
}
r = requests.get(url, headers = headers)

r.encoding = 'utf-8'
print(r.text[0:600])
```
