### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find('em').text,d.find_all('a')[-1].text)
```

### ③ 哪一個類別的文章最多

```python
import pandas as pd
item = soup.find(class_="part_list_2").find_all('h3')
time = [x.find(class_="date").text for x in item]
types = [x.find('em').text for x in item]
title = [x.find_all('a')[-1].text for x in item]
data = pd.DataFrame({'time': time, 'types':types, 'title':title})
data.groupby('types').size()
```
